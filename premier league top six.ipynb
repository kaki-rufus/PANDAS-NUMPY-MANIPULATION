```python
player_offsides = df['Offsides']
max_offsides = player_offsides.max()
max_offsides_player = player_offsides.idxmax()  # Find the index of the player with maximum offsides
player_name = df.loc[max_offsides_player, 'Name']  # Get the name of the player using the index

print(f"The player caught in offside position most times is: {player_name} with {max_offsides} in total")
```

```python
# Calculate total goals scored by players from each nationality
total_goals_by_nationality = df.groupby('Nationality')['Goals'].sum().sort_values(ascending=False)
print("Total goals scored by players from each nationality:")
print(total_goals_by_nationality)
```

```python
club_appearances = df.groupby('Club')['Appearances'].sum()
most_appearances_club = club_appearances.idxmax()
win_percentage = df.loc[df['Club'] == most_appearances_club, 'Wins'].sum() / club_appearances[most_appearances_club] * 100
print("Club with the most appearances:", most_appearances_club)
print("Win Percentage:", win_percentage)
```
